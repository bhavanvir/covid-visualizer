```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/sample_data/diabetes_prediction_dataset.csv')
df.head()
```

```python
df.describe()
```

```python
df.corr()
```

```python
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
sns.displot(df, x="bmi", bins=20)
plt.title('BMI Distribution')
plt.show()
```

```python
# Remove occurences of "Other" genders to colour the plot properly
df = df[df.gender != 'Other']

sns.violinplot(data=df, x='diabetes', y='bmi', hue='gender', split=True)
plt.title('BMI vs. Diabetes split by Gender')
plt.show()
```
